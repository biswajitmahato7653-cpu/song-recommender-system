```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
df = pd.read_csv(r"C:\Users\Dell\OneDrive\Music\Hindi_Songs_Dataset.csv")
print(df.head())
```

```python
df['combined'] = (
    df['song_name'].astype(str) + " " +
    df['artist'].astype(str) + " " +
    df['album'].astype(str) + " " +
    df['genre'].astype(str)
)
```

```python
df['combined']
```

```python
tfidf = TfidfVectorizer(stop_words='english')
vectors = tfidf.fit_transform(df['combined'])
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(song):
    song=song.lower()

    try:
        index=df[df['song_name'].str.lower()== song].index[0]
    except:
        return []
    distance=similarity[index]
    songs_list=sorted(list(enumerate(distance)),
                      reverse=True,key=lambda x:x[1])[1:6]
    recommended_songs=[]
    for i in songs_list:
        recommended_songs.append(df.iloc[i[0]].song_name)
    return recommended_songs
```

```python
pickle.dump(similarity,open('similarity.pkl','wb'))
pickle.dump(df.to_dict(),open('songs_dict.pkl','wb'))
pickle.dump(recommend,open('recommend.pkl','wb'))
```
